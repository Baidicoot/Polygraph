```python
from tree_generation import TreeGenerator
from tree import DialogueTree
```

```python
generator = TreeGenerator()
```

```python
TREE_COUNT = 9
HONESTY = "mildly_deceptive"

for i in range(8, TREE_COUNT):
    # Catch errors and go to next tree
    try:
        tree = generator.generate_tree(
            num_turns=5, 
            branches=3, 
            tree_id=i, 
            data_dir="labels", 
            honesty=HONESTY, 
            model="gpt-3.5-turbo", 
            temperature=1.15,
            include_success_likelihood=True
        )
        tree.root.calculate_success_pct()

    except Exception as e:
        print(f"Tree {i} failed with error: {e}")
        continue

    # Save the tree
    success_pct = tree.get_tree_success_pct()
    success_likelihood = tree.get_tree_success_likelihood()
    print(f"Tree {i} success pct: {success_pct}")
    print(f"Tree {i} success likelihood: {success_likelihood}")
    print(f"Saving tree {i}...")
    tree.save_as_json(f"dialogues_json/{HONESTY}-tree-{i}.json")
    #tree = DialogueTree()
    #tree.build_from_json(f"dialogues_json/mildly_deceptive-tree-{i}.json")
    tree.traverse_and_save_text(path="dialogues_text", include_final_user_response=True)
```
